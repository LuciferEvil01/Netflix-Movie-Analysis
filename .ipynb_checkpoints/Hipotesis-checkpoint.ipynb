```python
# Importar las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import chisquare
from scipy.stats import shapiro, levene, mannwhitneyu
from scipy.stats import pearsonr, spearmanr
import seaborn as sns

from statsmodels.stats.proportion import proportions_ztest
```

## 1. Carga y Preprocesamiento de Datos

En esta sección, cargamos el conjunto de datos de Netflix y realizamos el preprocesamiento necesario para el análisis:

- Cargamos el archivo CSV `netflix_titles.csv`.
- Filtramos  las películas (`type == 'Movie'`) y las series (`type == 'TV Show'`).
- Limpiamos y convertimos la columna `duration` a formato numérico.
- Eliminamos filas con valores faltantes en las columnas relevantes.


```python
# Filtrar solo las películas
df = pd.read_csv('netflix_titles.csv')
movies = df[df['type'] == 'Movie']
shows = df[df['type'] == 'TV Show']

# Limpiar y convertir la columna 'duration' a numérico
movies.loc[:, 'duration'] = movies['duration'].str.replace(' min', '')
movies.loc[:, 'duration']= pd.to_numeric(movies['duration'], errors='coerce')

shows.loc[:, 'duration'] = shows['duration'].str.replace(' Seasons', '')
shows.loc[:, 'duration'] = shows['duration'].str.replace(' Season', '')
shows.loc[:, 'duration']= pd.to_numeric(shows['duration'], errors='coerce')

# Eliminar filas con valores NaN en 'duration'
movies=movies.dropna(subset=['duration'])
shows=shows.dropna(subset=['duration'])

# Mostrar las primeras filas del DataFrame de películas
movies.head()
```

## 2. Estadísticos Descriptivos Básicos

Calculamos algunos estadísticos descriptivos para entender mejor nuestros datos:

- Cantidad total de películas y series.
- Duración mínima y máxima.
- Media y varianza de la duración.


```python
# Cantidad de películas
total_movies = len(movies)
print(f"Cantidad de películas: {total_movies}")

# Duración mínima y máxima
duration_min = movies['duration'].min()
movie_min = movies[movies['duration'] == duration_min]['title'].values
print(f"Duración mínima: {duration_min} minutos")
print(f"Películas con duración mínima: {movie_min}")

duration_max = movies['duration'].max()
movie_max = movies[movies['duration'] == duration_max]['title'].values
print(f"Duración máxima: {duration_max} minutos")
print(f"Películas con duración máxima: {movie_max}")

# Media de la duración
media_movie = movies['duration'].mean()
print(f"Media de duración de las películas: {media_movie:.2f} minutos")

# Varianza muestral de la duración
varianza_movie = movies['duration'].var(ddof=1)
print(f"Varianza muestral de las películas: {varianza_movie:.2f} minutos^2")
```

```python
# Cantidad de series
total_shows = len(shows)
print(f"Cantidad de series: {total_shows}")

# Duración máxima
duration_max = shows['duration'].max()
show_max = shows[shows['duration'] == duration_max]['title'].values
print(f"Duración máxima  : {duration_max} temporadas")
print(f"Serie con duración máxima: {show_max}")

# Media de la duración
media_show = shows['duration'].mean()
print(f"Media de duración de las series: {media_show:.2f} temporadas")

# Varianza muestral de la duración
varianza_show = shows['duration'].var(ddof=1)
print(f"Varianza muestral de las series: {varianza_show:.2f} ")
```

## 3. Cálculo de Intervalos de Confianza

Calculamos los intervalos de confianza al 95% para la media y la varianza de la duración de las películas.

- **Intervalo de confianza para la media**: Utilizamos la distribución t de Student.
- **Intervalo de confianza para la varianza**: Utilizamos la distribución chi-cuadrado.


```python
# Nivel de confianza
conf = 0.95
alpha = 1 - conf
n = len(movies['duration'])
grados_libertad = n - 1

# Error estándar
sem_movies = stats.sem(movies['duration'])

# Intervalo de confianza para la media
media_interval = stats.t.interval(conf, grados_libertad, loc=media_movie, scale=sem_movies)
print(f"Intervalo de confianza al {conf*100}% para la media: {media_interval}")

# Valores críticos chi-cuadrado
chi2_inf = stats.chi2.ppf(alpha/2, grados_libertad)
chi2_sup = stats.chi2.ppf(1 - alpha/2, grados_libertad)

# Intervalo de confianza para la varianza
intervalo_varianza = ((grados_libertad)*varianza_movie/chi2_sup, (grados_libertad)*varianza_movie/chi2_inf)
print(f"Intervalo de confianza al {conf*100}% para la varianza: {intervalo_varianza}")
```

## 4. Pruebas de Hipótesis

Realizamos varias pruebas de hipótesis para analizar nuestras datos.

### 4.1. Prueba de Hipótesis para la Media

**Objetivo**: Comprobar si la media de las duraciones de las películas es mayor a 90 minutos.

- **Hipótesis nula (\( H_0 \))**: La media es menor o igual a 90 minutos (\( \mu \leq 90 \)).
- **Hipótesis alternativa (\( H_a \))**: La media es mayor a 90 minutos (\( \mu > 90 \)).


```python
# Valor hipotético de la media
mu_0 = 90

# Estadístico t
t_stat = (media_movie - mu_0) / sem_movies
p_value = 1 - stats.t.cdf(t_stat, grados_libertad)

print(f"Estadístico t: {t_stat:.4f}")
print(f"Valor p: {p_value:.4f}")

if p_value < alpha:
    print("Se rechaza la hipótesis nula: la media de las duraciones es mayor a 90 minutos.")
else:
    print("No se puede rechazar la hipótesis nula: no hay evidencia suficiente para afirmar que la media es mayor a 90 minutos.")
```

### 4.2. Prueba de Hipótesis para la Varianza

**Objetivo**: Comprobar si la varianza de las duraciones de las películas es igual a 800 minutos².

- **Hipótesis nula (\( H_0 \))**: La varianza es igual a 800 minutos² (\( \sigma^2 = 800 \)).
- **Hipótesis alternativa (\( H_a \))**: La varianza es diferente de 800 minutos² (\( \sigma^2 \ne 800 \)).


```python
# Varianza hipotética
sigma0_squared = 800

# Estadístico chi-cuadrado
chi2_stat = (grados_libertad * varianza_movie) / sigma0_squared

# Valores críticos chi-cuadrado (usados anteriormente)
# chi2_inf y chi2_sup ya calculados

print(f"Estadístico chi-cuadrado: {chi2_stat:.4f}")
print(f"Valores críticos chi-cuadrado: {chi2_inf:.4f}, {chi2_sup:.4f}")

if chi2_stat < chi2_inf or chi2_stat > chi2_sup:
    print("Se rechaza la hipótesis nula: la varianza de las duraciones es diferente de 800 minutos².")
else:
    print("No se puede rechazar la hipótesis nula: no hay evidencia suficiente para afirmar que la varianza difiere de 800 minutos².")
```

### 4.3. Prueba de Hipótesis para la Proporción

**Objetivo**: Comprobar si al menos el 70% de las películas tienen una duración mayor a 80 minutos.

- **Hipótesis nula (\( H_0 \))**: La proporción es menor o igual a 0.70 (\( p \leq 0.70 \)).
- **Hipótesis alternativa (\( H_a \))**: La proporción es mayor a 0.70 (\( p > 0.70 \)).


```python
from statsmodels.stats.proportion import proportions_ztest

# Datos
n_exitos = (movies['duration'] > 80).sum()
p_muestral = n_exitos / n
p0 = 0.70

# Prueba de proporción
stat, p_value = proportions_ztest(count=n_exitos, nobs=n, value=p0, alternative='larger')

print(f"Proporción muestral: {p_muestral:.4f}")
print(f"Estadístico Z: {stat:.4f}")
print(f"Valor p: {p_value:.4f}")

if p_value < alpha:
    print("Se rechaza la hipótesis nula: más del 70% de las películas tienen duración mayor a 80 minutos.")
else:
    print("No se puede rechazar la hipótesis nula: no hay evidencia suficiente para afirmar que más del 70% de las películas tienen duración mayor a 80 minutos.")
```

## 5. Categorías más comunes en películas y series

**Objetivo 1 **: Determinar si la categoría más común de las películas. es "International Movies".
- **Hipótesis nula (\( H_0 \))**: La categoría "International Movies" no es la categoría más común entre las películas; su frecuencia no difiere significativamente de las demás categorías.
- **Hipótesis alternativa (\( H_a \))**: La categoría "International Movies" es la categoría más común entre las películas; su frecuencia es significativamente mayor que la de otras categorías.


```python

# Manejar valores faltantes
movies['listed_in'] = movies['listed_in'].fillna('')

# Separar las categorías y expandir el DataFrame
movies['category_list'] = movies['listed_in'].apply(lambda x: x.split(', '))
movies_exploded = movies.explode('category_list')

# Contar las frecuencias de cada categoría
category_counts = movies_exploded['category_list'].value_counts()

# Mostrar las 10 categorías más comunes
print("Frecuencias Observadas de las Categorías en Películas:")
print(category_counts.head(10))

# **2. Calcular las Frecuencias Esperadas**

# Seleccionar las 10 categorías más comunes para el análisis
top_categories = category_counts.nlargest(10)
observed_frequencies = top_categories.values
categories = top_categories.index.tolist()

# Calcular las frecuencias esperadas asumiendo distribución uniforme
total_observations = observed_frequencies.sum()
expected_frequency = total_observations / len(observed_frequencies)
expected_frequencies = [expected_frequency] * len(observed_frequencies)

# **3. Prueba Chi-cuadrado de Bondad de Ajuste**

chi2_stat, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

print(f"\nEstadístico Chi-cuadrado: {chi2_stat:.2f}")
print(f"Valor p: {p_value:.4e}")

# **4. Interpretación de los Resultados**

alpha = 0.05
if p_value < alpha:
    print("\nConclusión: Se rechaza la hipótesis nula. Existe una diferencia significativa en las frecuencias de las categorías.")
    # Identificar si "Dramas" es la categoría más frecuente
    if categories[0] == 'International Movies':
        print("La categoría 'International Movies' es significativamente más común que otras categorías en películas.")
    else:
        print(f"La categoría más común es '{categories[0]}', no es 'Dramas'.")
else:
    print("\nConclusión: No se puede rechazar la hipótesis nula. No hay evidencia suficiente para afirmar que las frecuencias difieren.")
```

## 6. Rating en películas y serie

**Objetivo 1 **: Determinar si el rating  más común de las películas de EE.UU. es "TV-MA".
- **Hipótesis nula (\( H_0 \))**: El rating más común entre las películas de EE.UU. no es TV-MA.
- **Hipótesis alternativa (\( H_a \))**: El rating más común entre las películas de EE.UU. es TV-MA.


```python
# Manejar valores faltantes y filtrar películas de EE.UU.
df['country'] = df['country'].fillna('')
df['rating'] = df['rating'].fillna('')

us_movies = df[(df['country'].str.contains('United States')) & (df['type'] == 'Movie')]
us_movies.dropna(subset=['country'], inplace=True)
us_tvshow=df[(df['country'].str.contains('United States')) & (df['type'] == 'TV Show')]
us_tvshow.dropna(subset=['country'], inplace=True)

# Contar el número total de películas de EE.UU.
total_us_movies = len(us_movies)
print(f"Total de películas de EE.UU.: {total_us_movies}")
# Contar las frecuencias de ratings en películas de EE.UU.
rating_counts = us_movies['rating'].value_counts()
print("\nFrecuencias de ratings en películas de EE.UU.:")
print(rating_counts)

# **2. Análisis de Frecuencias**

# Identificar el rating más común
most_common_rating = rating_counts.index[0]
most_common_count = rating_counts.iloc[0]

# Verificar si el rating más común es TV-MA
if most_common_rating == 'TV-MA':
    print(f"\nEl rating más común de la muestra  es TV-MA con {most_common_count} películas.")
else:
    print(f"\nEl rating más común es {most_common_rating} con {most_common_count} películas.")
    print("Por lo tanto, no se puede rechazar la hipótesis nula.")
    # Fin del análisis si TV-MA no es el rating más común
    exit()

# **3. Prueba de Hipótesis**

# Identificar el segundo rating más común
second_common_rating = rating_counts.index[1]
second_common_count = rating_counts.iloc[1]

# Números de éxitos y observaciones
count = np.array([most_common_count, second_common_count])
nobs = np.array([total_us_movies, total_us_movies])

# Prueba de proporciones (unilateral hacia la derecha)
stat, p_value = proportions_ztest(count, nobs, alternative='larger')

print(f"\nEstadístico Z: {stat:.4f}")
print(f"Valor p: {p_value:.4f}")

# **4. Interpretación de Resultados**

alpha = 0.05
if p_value < alpha:
    print("\nConclusión: Se rechaza la hipótesis nula. El rating más común en EE.UU. es TV-MA.")
else:
    print("\nConclusión: No se puede rechazar la hipótesis nula. No hay evidencia suficiente para afirmar que TV-MA es el rating más común en EE.UU.")





total_us_tvshow = len(us_tvshow)
print(f"Total de series de EE.UU.: {total_us_tvshow}")
# Contar las frecuencias de ratings en series de EE.UU.
rating_counts = us_tvshow['rating'].value_counts()
print("\nFrecuencias de ratings en series de EE.UU.:")
print(rating_counts)

# **2. Análisis de Frecuencias**

# Identificar el rating más común
most_common_rating = rating_counts.index[0]
most_common_count = rating_counts.iloc[0]

# Verificar si el rating más común es TV-MA
if most_common_rating == 'TV-MA':
    print(f"\nEl rating más común de la muestra  es TV-MA con {most_common_count} series.")
else:
    print(f"\nEl rating más común es {most_common_rating} con {most_common_count} series.")
    print("Por lo tanto, no se puede rechazar la hipótesis nula.")
    # Fin del análisis si TV-MA no es el rating más común
    exit()

# **3. Prueba de Hipótesis**

# Identificar el segundo rating más común
second_common_rating = rating_counts.index[1]
second_common_count = rating_counts.iloc[1]


# Números de éxitos y observaciones
count = np.array([most_common_count, second_common_count])
nobs = np.array([total_us_tvshow, total_us_tvshow])

# Prueba de proporciones (unilateral hacia la derecha)
stat, p_value = proportions_ztest(count, nobs, alternative='larger')

print(f"\nEstadístico Z: {stat:.4f}")
print(f"Valor p: {p_value:.4f}")

# **4. Interpretación de Resultados**

alpha = 0.05
if p_value < alpha:
    print("\nConclusión: Se rechaza la hipótesis nula. El rating más común en EE.UU. es TV-MA.")
else:
    print("\nConclusión: No se puede rechazar la hipótesis nula. No hay evidencia suficiente para afirmar que TV-MA es el rating más común en EE.UU.")
```

## 5. Análisis Comparativo entre Películas de EE.UU. y Otros Países

**Objetivo**: Determinar si existe una diferencia significativa en la duración media de las películas producidas en Estados Unidos y las de otros países.

### 5.1. Preprocesamiento de la Columna `country`

Como las películas pueden tener múltiples países asociados, separamos los países y expandimos las filas para que cada país tenga su propia fila.


```python
# Eliminar filas con valores NaN en 'country'
movies.dropna(subset=['country'], inplace=True)

# Separar los países y expandir las filas
movies['country'] = movies['country'].str.split(', ')
movies = movies.explode('country')

# Asegurarse de que no hay valores nulos
movies.dropna(subset=['duration', 'country'], inplace=True)

# Verificar los datos
movies[['title', 'duration', 'country']].head()
```

```python
# Separar las películas de EE.UU. y otros países
movies_usa = movies[movies['country'].str.contains('United States', na=False)]
movies_others = movies[~movies['country'].str.contains('United States', na=False)]

movies_usa.loc[:,'duration']  = pd.to_numeric(movies_usa['duration'], errors='coerce')
movies_others.loc[:,'duration'] = pd.to_numeric(movies_others['duration'], errors='coerce')

movies_usa['duration'] = movies_usa['duration'].astype(int)
movies_others['duration'] = movies_others['duration'].astype(int)
# Duraciones
durations_usa = movies_usa['duration']
durations_others = movies_others['duration']
```

### 5.2. Prueba T de Dos Muestras

**Hipótesis:**

- **Hipótesis nula (\( H_0 \))**: Las duraciones medias son iguales (\( \mu_{\text{USA}} = \mu_{\text{Otros}} \)).
- **Hipótesis alternativa (\( H_a \))**: Las duraciones medias son diferentes (\( \mu_{\text{USA}} \ne \mu_{\text{Otros}} \)).


```python
# Prueba T de dos muestras independientes
t_stat, p_value = stats.ttest_ind(durations_usa, durations_others, equal_var=False)

if p_value < alpha:
    print("Se rechaza la hipótesis nula: las duraciones medias son diferentes.")
else:
    print("No se puede rechazar la hipótesis nula: no hay evidencia suficiente para decir que las duraciones medias son diferentes.")
```

### 6. Cómo obtener intervalos de confianza y pruebas de hipótesis para variables no normales.

Cuando las variables no siguen una distribución normal, las pruebas paramétricas estándar pueden no ser adecuadas. Aquí hay algunas estrategias:

#### a) Transformaciones de datos:

- **Logarítmica, raíz cuadrada o inversa para aproximar la normalidad.**
- Útil si los datos están sesgados positivamente.

#### b) Métodos no paramétricos:

**Intervalos de confianza mediante bootstrap:**

- Re muestreo con reemplazo para crear múltiples muestras simuladas.
- Calcular la estadística de interés en cada muestra.
- Construir el intervalo a partir de percentiles de estas distribuciones.

**Pruebas de hipótesis no paramétricas:**

- **Prueba de Wilcoxon** para comparar medianas.
- **Prueba de Kruskal-Wallis** para más de dos grupos.

#### c) Distribuciones alternativas:

- Si conoces la distribución subyacente (por ejemplo, binomial, Poisson), utiliza métodos específicos para esas distribuciones.

### Analisis 2: Correlación entre duration y rating
#### Pregunta :¿La clasificación por edad está relacionada con la duración de las películas?
#### Motivacion :Las películas para audiencias adultas podrían tener duraciones distintas a las de público infantil.


```python
# Mapear ratings a valores numéricos
rating_levels = {
    'G': 1,
    'TV-G': 1,
    'PG': 2,
    'TV-PG': 2,
    'PG-13': 3,
    'TV-14': 3,
    'R': 4,
    'TV-MA': 4,
    'NC-17': 5,
    'NR': np.nan,
    'UR': np.nan,
    'TV-Y': 1,
    'TV-Y7': 1,
    'TV-Y7-FV': 1,
    'TV-NR': np.nan,
    'TV-UNRATED': np.nan,
    'UNRATED': np.nan,
    '': np.nan
}

movies.loc[:,'rating_num'] = movies['rating'].map(rating_levels)

# Eliminar filas con valores faltantes en 'rating_num'
movies_rating = movies.dropna(subset=['rating_num', 'duration_num'])
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='rating_num', y='duration_num', data=movies_rating)
plt.title('Duración vs Rating (Películas)')
plt.xlabel('Rating Numérico')
plt.ylabel('Duración (minutos)')
plt.show()
```

```python
# Prueba de correlación de Spearman
corr_coef, p_value = spearmanr(movies_rating['rating_num'], movies_rating['duration_num'])

print(f"Coeficiente de correlación de Spearman: {corr_coef:.4f}")
print(f"Valor p: {p_value:.4e}")
```

#### Interpretacion
 **Coeficiente de correlacion**:
- Valores cercanos a 1 o -1 indican una relación lineal fuerte.
- Valores cercanos a 0 indican poca o ninguna relación lineal.
 
  **Valor p**:
- i el valor p es menor que el nivel de significancia, la correlación es estadísticamente significativa.

### Analisis 3: Correlación entre release_year y date_added
#### Pregunta :¿Hay una relación entre el año de lanzamiento de un título y la fecha en que fue agregado a Netflix?
#### Motivacion :Es posible que Netflix agregue más rápidamente títulos recientes que antiguos.


```python
# Eliminar filas con valores faltantes en 'date_added'
df_date = df.dropna(subset=['date_added', 'release_year'])

# Convertir 'date_added' a número de días desde la fecha mínima
df_date.loc[:,'days_since_added'] = (df_date['date_added'] - df_date['date_added'].min()).dt.days

# Variables para análisis
x = df_date['release_year']
y = df_date['days_since_added']
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x, y=y)
plt.title('Fecha Agregada vs Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Días desde que se agregó a Netflix')
plt.show()
```

```python
# Prueba de correlación de Pearson
corr_coef, p_value = pearsonr(x, y)

print(f"Coeficiente de correlación de Pearson: {corr_coef:.4f}")
print(f"Valor p: {p_value:.4e}")
```

#### Interpretacion
 **Coeficiente de correlacion**:
- Valores cercanos a 1 o -1 indican una relación lineal fuerte.
- Valores cercanos a 0 indican poca o ninguna relación lineal.
 
  **Valor p**:
- i el valor p es menor que el nivel de significancia, la correlación es estadísticamente significativa.

### Matriz de Correlacion


```python
# Filtrar películas
# Mapear ratings a niveles numéricos
rating_levels = {
    'TV-Y': 1,     # Todos los niños
    'TV-Y7': 2,    # Niños de 7 años o más
    'TV-Y7-FV': 3, # Niños de 7 años o más con violencia fantástica
    'G': 4,        # Público general
    'TV-G': 5,     # Público general
    'PG': 6,       # Guía paternal sugerida
    'PG-13': 7,    # Guía paternal fuerte sugerida, no apto para menores de 13
    'TV-PG': 8,    # Guía parental sugerida
    'TV-14': 9,    # No apto para menores de 14
    'R': 10,        # Restringido, menores de 17 requieren acompañamiento
    'TV-MA': 11,    # Solo para adultos
    'NC-17': 12,    # No apto para menores de 17
    'NR': 13,  # No clasificado
    'UR': 14,  # Sin clasificación
    'UNRATED': np.nan,  # Sin clasificación
    '': np.nan     # Sin datos
}

df['rating_num'] = df['rating'].map(rating_levels)



movies = df[df['is_movie'] == 1].copy()

# Eliminar filas con valores faltantes en variables numéricas
movies = movies.dropna(subset=['duration_num', 'release_year', 'days_since_added', 'rating_num'])

# Seleccionar variables numéricas
numeric_vars_movies = ['duration_num', 'release_year', 'days_since_added', 'rating_num']
# Calcular matriz de correlación
corr_matrix_movies = movies[numeric_vars_movies].corr()

# Mostrar la matriz de correlación
print("Matriz de Correlación (Películas):")
print(corr_matrix_movies)

# Visualización con seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix_movies, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación - Películas')
#plt.savefig('matriz_correlacion_peliculas.png', format='png', dpi=300)
plt.show()
```

```python
# Filtrar series
series = df[df['is_movie'] == 0].copy()

# Eliminar filas con valores faltantes en variables numéricas
series = series.dropna(subset=['seasons_num', 'release_year', 'days_since_added', 'rating_num'])

# Seleccionar variables numéricas
numeric_vars_series = ['seasons_num', 'release_year', 'days_since_added', 'rating_num']

# Calcular matriz de correlación
corr_matrix_series = series[numeric_vars_series].corr()

# Mostrar la matriz de correlación
print("Matriz de Correlación (Series):")
print(corr_matrix_series)

# Visualización con seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix_series, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación - Series')
#plt.savefig('matriz_correlacion_series.png', format='png', dpi=300)
plt.show()

```
